```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = {
'CZ 75': np.random.normal(60, 5, 12),
'Desert Eagle': np.random.normal(76, 6, 12),
'Sig Seuer P226': np.random.normal(40, 4, 12),
'Beretta 92': np.random.normal(55, 7, 12),
'Smith & Wesson M&P': np.random.normal(47, 16, 12),
'Walther': np.random.normal(100, 23, 12),
'Ruger SR9': np.random.normal(47, 5, 12),
'Glock 17': np.random.normal(101, 15, 12),
'Springfield XD': np.random.normal(80, 4, 12),
"Heckler & Koch USP": np.random.normal(70, 13, 12),
}
df = pd.DataFrame(data).transpose()
df
```

```python
ax = sns.lineplot(df)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1))
```

```python
p0 = df.sum() / df.shape[0]
p0
```

```python
std = ((df - p0) ** 2).sum() / (df.shape[0] - 1)
std = std ** (1/2)
print(std)
```

```python
x_extrapol = p0 + np.random.normal(0, std, len(p0))
pd.concat([df, pd.DataFrame([x_extrapol], columns = x_extrapol.index)]).reset_index(drop = True)
```

```python
reliability = std/p0
print(reliability)
```

```python
cond1 = ((df - p0) < 2 * std).all()
cond1
```

```python
cond2 = p0 > 2 * std
cond2
```

```python
cond3 = (df > 0).all()
cond3
```

```python
product_color = pd.Series(dtype = 'string')
for name in df.columns:
    if (not cond3[name]):
        product_color[name] = 'Красный'
    elif (not cond1[name] and not cond2[name]):
        product_color[name] = 'Оранжевый'
    elif (not cond1[name] or not cond2[name]):
        product_color[name] = 'Жёлтый'
    else:
        product_color[name] = 'Зелёный'
product_color
```
